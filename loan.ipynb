```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = [
    'person_age',
    'person_income',
    'loan_amnt',
    'loan_int_rate',
    'credit_score'
]

plt.figure(figsize=(15,8))

for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)   # 2 rows, 3 columns
    sns.histplot(df[col], bins=40, kde=True)
    plt.title(f"{col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = [
    'person_age',
    'person_income',
    'loan_amnt',
    'loan_int_rate',
    'credit_score'
]

fig, axes = plt.subplots(2, 3, figsize=(18, 8))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.boxplot(x=df[col], ax=axes[i])
    axes[i].set_title(col, fontsize=12)
    axes[i].set_xlabel("")

# Remove the empty subplot (6th one)
fig.delaxes(axes[-1])

plt.suptitle("Outlier Detection Using Boxplots", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cat_cols = [
    'person_gender',
    'person_education',
    'person_home_ownership',
    'loan_intent'
]

fig, axes = plt.subplots(2, 2, figsize=(18, 10))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    sns.countplot(y=col, data=df, ax=axes[i])
    axes[i].set_title(f"Distribution of {col}", fontsize=12)
    axes[i].set_xlabel("Count")
    axes[i].set_ylabel(col)

plt.suptitle("Distribution of Categorical Features", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(
    df[num_cols + ['loan_status']].corr(),
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)
plt.title("Correlation Heatmap")
plt.show()
```

```python

df['log_income'] = np.log1p(df['person_income'])

plt.figure(figsize=(8,5))
sns.boxplot(x='loan_status', y='log_income', data=df)
plt.title("Log Income vs Loan Approval Status")
plt.xlabel("Loan Status")
plt.ylabel("Log(Income)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='person_income',
    y='loan_percent_income',
    hue='loan_status',
    data=df,
    alpha=0.4
)
plt.title("Income vs Loan-to-Income Ratio by Loan Status")
plt.xlabel("Income")
plt.ylabel("Loan % of Income")
plt.show()
```
